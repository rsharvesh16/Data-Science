```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/carprices.csv')
df
```

```python
df.rename(columns={'Car Model':'Car','Sell Price($)':'Sell_Price','Age(yrs)':'Age'},inplace=True)
```

```python
df
```

```python
dummies=pd.get_dummies(df.Car)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
merged.rename(columns={'Audi A5':'Audi'},inplace=True)
merged
```

```python
final=merged.drop(['Car','Audi'],axis='columns')
final
```

```python
reg=linear_model.LinearRegression()
reg
```

```python
X=final.drop('Sell_Price',axis='columns')
X
```

```python
y=final.Sell_Price
y
```

```python
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
reg.predict([[45000,4,0,1]])
```
